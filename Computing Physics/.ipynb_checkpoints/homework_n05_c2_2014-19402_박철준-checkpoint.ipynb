## ** Optimization in Machine Learning ** 

### ** 경사하강법 (Gradient Descent Method) ** 


임의의 신경망 모형이 있다고 하자. 
즉, 신경망의 구조과 신호처리방식이 구체적으로 정의되어 있어서 여러가지 하이퍼파라메터 h들과 그에 따른 연결 가중치 학습파라메터$\vec{w}$가 구체화되어 어떤값으로 초기화되어 있음을 의미한다. **[최하단의 참조자료2 참조]**

이때 목적값(label, 정답라벨, 지도라벨, $y_i$)과 함께 주어진 입력 데이터 하나 $x_{i}$ 에 대하여 그 출력이 $f(x_i;\vec{w},h)$이라면, 
이에 대한 오차함수 $E$ (error/loss/cost/objective function)를

> $E_i(\vec{w};x_i) = \left[\,f(x_i;\vec{w})-y_i\,\right]^2$

과 같은 제곱오차(SE,Squared Error) 형식으로 정의할 수 있으며, 샘플의 크기(갯수)가 $n$인 모든 데이터($i=1..n$)에 대한 평균오차는 평균제곱오차(MSE,Mean Squared Error) 형식으로 다음과 같이 정의될 수 있다

> $E^{(n)}(\vec{w};\{x_1,...,x_n\}) = \frac{1}{n}\sum_{i=1}^n E_i $

이렇게 여러가지 크기($n$)로 주어진 데이터에 대하여, 현재 기계모형이 가지는 평균오차는 연결가중치 파라메터 $\vec{w}$의 함수로서,  
현재의 가중치값들은, 현 위치에서 오차함수값이 가장 빠르게 줄어드는 방향으로 적절한 학습률 파라메터 (learning rate, $\eta$)만큼의  스케일 팩터를 가지고 $\delta \vec{w} = -\eta\vec{\nabla}_{w}E(\vec{w})$) 만큼의 변화를 갖도록 다음과 같이 갱신한다.

> $\vec{w}_{k+1} = \vec{w}_{k} - \eta \vec{\nabla}_{w} E^{(n)}(\vec{w}_k;x)$

상기한 방법을 사용하여 주어진 데이터 입력에 대해 현재 모형이 가진 오차함수값를 줄이는 방향으로, 신경망 가중치의 갱신(학습)을 구현하게되며,  
이렇게 어떤 목적함수(objective function, 최소화의 대상이 되는 함수)의 국소적인 경사도(gradient)정보를 사용하여 
목적함수를  최적화하는 알고리즘을, 경사하강법 (Gradient Descent Method)이라고 한다.

현재 딥러닝에서 연결 가중치값의 최적화(학습)에는, 목적함수의 지형정보(gradient, Hessian, (differential) geometry)에 기반을 두고 위치를 갱신해나가는 최적화 알고리즘 (gradient-based optimization algorithms)들이 주로 사용되고 있으며, 위의 기본적인 경사하강법을 보완하여 보다 향상된 성능을 보여주는 최적화 알고리즘들도 널리 사용되고 있다.

#### ** 경사하강법 예시) ** 
주어진 어떤 데이터에 대해 가중치 공간에서 오차함수가 다음과 같이 정의되어있다.  
경사하강법을 사용하여 오차함수가 최소가 되는 위치(극소값/local miniumum point/minimizer)를 탐사해보자.

> $E(w) = w^2$ ,  ($w_0 = 2$)

> $E(w) = w^4 - 3w^3 + 2$ , ($w_0 = 2$)

*학습률을 조정해가면서 갱신되는 위치의 궤적을 시각화해보자.

```python
import numpy as np

E = lambda w: w**2
dE = lambda w: 2.*w

# E = lambda w: w**4 - 3.*w**3 + 2.
# dE = lambda w: 4.*w**3 - 9.*w**2

# learning rate  
eta = 0.8

# initial position
w_now = 2.0
w_prev = None
w_list = [w_now]

# iteration control
Niter_max = 1000
Niter = 0 

# convergence check
# - tolerance on step size, dw 
# - status converged to a local minimum, if dw < dw_conv
dw_conv = 0.0001
dw = dw_conv + 1.

for i in range(Niter_max):
    print (' * Niter: %5i -> w: %10.8f'%(i,w_now))
    w_prev = w_now
    dw = eta*dE(w_prev)
    if abs(dw) < dw_conv:
        break        
    w_now -= dw
    w_list.append(w_now)
```

```python
import matplotlib.pyplot as plt

w_np = np.arange(-2.5,2.5,0.01)

fig, axes = plt.subplots(nrows=1,ncols=1,figsize=(6,4))

axes.plot(w_list,E(np.array(w_list)), marker='.', markersize=10, lw=1, ls='--')
axes.plot(w_np,E(w_np),color='black')
axes.set_xlabel('w')
axes.set_ylabel('E(w)')

axes.grid(True)
```

### **정규과제05 문제1) ** 
주어진 어떤 데이터에 대해, 2차원의 가중치 공간에서 오차함수가 다음과 같이 정의되어 있다 [밑의 첨부 플랏 참조]  
경사하강법을 사용하여 오차함수가 최소가 되는 위치(local miniumum point)를 탐색 및 탐사해보자. [80pt]

> $E(w_1, w_2) = (w_1 + w_2) + \frac{1}{\mu} (w_1^2 + w_2^2 - 1)^2 $    

> with $w_0 = (0,1)$ & $\mu=0.05$

1) 경사하강법을 2차원 공간에서 구현하여, 학습률과 초기값을 조정해가면서 갱신되는 위치의 궤적을 matplotlib.pyplot.contour와 함께 시각화해보자. [20pt]

2) 극소점의 위치를 해석적으로 구해보자. [10pt]

3) 2)번의 극소점으로 더욱 빨리 수렴해나가는 알고리즘을 구현해보자 [50pt]

   예를 들면,  
   * 단순 경사하강법에서의 학습률 최적화를 시도해보거나,  
   * 단순 경사하강법 이외의 보다 발전된 알고리듬을 시도해볼 수도 있을 것이다. 

    [기본 30pt. 예상되는 극소점에 효율적으로 빨리 안착하는 시연결과들에 대해서는 +20pt]


```python
###### %matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('default')
plt.rcParams["axes.grid"] = False

mu = 0.05
w1_0, w2_0 = 0., 1.

E = lambda w1,w2 : w1+w2 + 1/mu * (w1**2+w2**2-1)**2
dE_w1 = lambda w1,w2 : 1+4*(w1**3+(w2**2-1)*w1)/mu
dE_w2 = lambda w1,w2 : 1+4*(w2**3+(w1**2-1)*w2)/mu

#learning rate
eta = 0.01

# initial position
w1_now, w2_now = w1_0, w2_0
w1_prev, w2_prev = None, None
w1_list, w2_list = [w1_now], [w2_now]

# iteration control
Niter_max = 10000
Niter = 0 

# convergence check
# - tolerance on step size, dw 
# - status converged to a local minimum, if dw < dw_conv
dw_conv = 0.00001
dw1 = dw_conv + 1.
dw2 = dw_conv + 1.

for i in range(Niter_max):
    w1_prev, w2_prev = w1_now, w2_now
    dw1 = eta*dE_w1(w1_prev,w2_prev)
    dw2 = eta*dE_w2(w1_prev,w2_prev)
    if (abs(dw1) < dw_conv) and (abs(dw2) < dw_conv):
        print (' * Niter: %5i -> w1: %10.8f, w2: %10.8f'%(i,w1_now,w2_now))
        break        
    w1_now -= dw1
    w2_now -= dw2
    w1_list.append(w1_now)
    w2_list.append(w2_now)

w1_np, w2_np = np.linspace(-1.2, 1.2, 100), np.linspace(-1.2, 1.2, 100)
W1, W2 = np.meshgrid(w1_np, w2_np)

fig, axes = plt.subplots(nrows=1,ncols=1,figsize=(5,5))
contour = axes.contour(W1,W2,E(W1,W2),30)
axes.plot(w1_0,w2_0,color='red',marker='.',ms=5)
axes.plot(w1_list,w2_list,color='cyan', marker='.', markersize=2, lw=1, ls='--')
axes.set_xlabel('$w_1$',fontsize=15)
axes.set_ylabel('$w_2$',fontsize=15)

# fig.colorbar(contour, pad=0.01)

plt.savefig('where_is_my_vacuum.png')
# plt.show()
```

2번 해답
> $\frac{\partial E}{\partial w_1}=0, \frac{\partial E}{\partial w_2}=0$

> 위의 두 식은 $x^3+px+q=0$ 꼴의 연립삼차방정식이 된다.

>$\begin{cases}
1+\frac{4}{\mu}(w_1^3+(w_2^2-1)w_1)=0&\cdots1)\\
1+\frac{4}{\mu}(w_2^3+(w_1^2-1)w_2)=0&\cdots2)
\end{cases}$

1)-2)를 하면,
>$(w_1-w_2)(w_1^2+w_2^2-1)=0$

이 된다. 그러면,
>$w_1=w_2 ,\, w_1^2+w_2^2=1$

이다. 이 때, $w_1^2+w_2^2=1$이면, 1,2번 식이 $1=0$이 되어 버리기 때문에 $w_1=w_2$만 해가 된다.
이를 대입하여 1번 식을 정리하면,
>$w_1^3-\frac{1}{2}w_1+\frac{\mu}{8}=0$

이 된다. 카르다노의 해법에 의해 이 방정식의 해는
>$w_1=\omega^k\root{3}\of{-\frac{\mu}{16}+\sqrt{(\frac{\mu}{16})^2-(\frac{1}{6})^3}}+\omega^{3-k}\root{3}\of{-\frac{\mu}{16}-\sqrt{(\frac{\mu}{16})^2-(\frac{1}{6})^3}},\,(k=0,1,2)$

> when, $\omega=\frac{-1+\sqrt{3}i}{2}$

$w_1=-0.71021,0.0062505,0.70396$

## ** Representation in Machine Learning ** 


기계학습이라는 데이터 기반 통계학습 (statistical learning) 알고리즘을 다음과 같이  
1. 표현(representation), 
2. 평가(evaluation), 
3. 최적화(optimization)

의 3요소로 나눌 때,

**'표현'**은 *우선*, 주어진 입력데이터($x$)에 대하여, 데이터에 존재하는 패턴을 학습할 기계모형의 출력값  $f(x)$ 이 최종적으로  
학습을 끝내고 대응(mapping)될, 희망 출력값/라벨값 $y(x)$의 정의와 형태를 지칭한다.   
  
>예를 들면 어떤 분류문제(classification)를 풀 때에는,  
많은 경우, 주어진 입력이 어느 클래스에 속하는 지에 대한 확률값을  
모형의 출력값과 희망 출력값으로 설정하게되며,  
    
>혹은 어떤 회귀문제(regression)를 풀 때에는,  
주어진 입력이 구체적으로 주어진 연속적인 값에 대응되고, 이를 피팅해내는 어떤 연속함수로서의 출력을 원하므로,  
고로 모형의 희망 출력값은 데이터에 대응될 연속값 그 자체로 설정하게 된다.  

**그리고 더 나아가** 표현이라는 요소는,  
**학습해낼 매우 복잡하고 비선형적인 모형 $f(x)$을, 구체적으로 정의한 방식 그 자체도 의미한다.** 

즉 모형의 입력과 출력 뿐만이 아니라, 이 관계모형을 구체적으로 구현하는 방식 역시, 기계학습모형의 '표현'에 매우 중요한 요소가 되며, 다른 말로 하자면 입력과 출력과정에서의 모형의 '표현'이 단순히 $x\rightarrow f(x)$가 아니라, 보다 자세히는 

> $x\rightarrow f(x;w,h)$ 

라는 것이다. 

**여기서 $h$와 $w$는 각각 모형의 하이퍼파라메터(hyper-parameter)와 학습파라메터(learning parameter)를 지칭**하는데,  
이들은 어떤 기계학습 모형 $f(x)$이 구체적으로 어떻게 구현되었는지에 따라 다르게 정의되는 모형의 내부 파라메터로서,  
그 자체로 기계학습 모형의 주요한 '표현'요소가 된다.

>예를 들면, 인공신경망(Neural Network)이라는 기계모형은,  
각각 비선형 출력을 가진 퍼셉트론(perceptron)간의 가중치 연결을 통하여,  
전체적인 출력값 $f(x;w,h)$이 어떤 복잡한 패턴을 표현할 수 있는 능력(capacity)을 가지게 되며,  
이때 인공신경망의 신경망 층(layer) 수, 한 층에서의 퍼셉트론 수, 각 퍼셉트론의 활성화 함수(activation function) 종류,   
신경망 가중치의 규제방법(regularization scheme) 등등, **신경망 연결구조와 신호처리 방식** 등을 정의하는 수많은 파라메터들이  
인공신경망 모형의 하이퍼파라메터가 된다.

이렇게 일반적인 기계학습 모형에는, 하이퍼파라메터(hyperparameter) h가 존재하게 되며,   
구체적인 h의 선택에 따라서, 그 모형이 가지는 세부적인 학습파라메터 w의 수가 명확하게 정의되고 이에 따라서  
모형의 표현능력(capacity)이 결정된다.

>예를 들면 신경망의 층 수를 증가시킬수록 신경망 연결수가 늘어나게되어, 연결 가중치 학습파라메터(w)    
역시 그 수가 늘어나게 되고, 증가된 학습파라메터의 자유도만큼, 신경망의 출력모형 역시 더욱 복잡한 패턴의 묘사가 가능하게된다.

**이러한 방식으로 구체화되어 표현된 어떤 기계모형 $f(x;w,h)$이 있을때,   
주어진 입력값 $x$과 더불어, 대응되어야하는 출력값 $y(x)$ 데이터에 대하여,**

> $x \rightarrow f(x;w,h)\rightarrow y(x)$

**가 되도록, (모든 데이터에 대하여) 학습파라메터 w를 조정(학습)해나가는 것이 바로 (지도학습으로서의) 기계학습이다.**
  
>   *풍부한 양질의 데이터*를 통해 *잘* 훈련된 기계학습모형  $f(x;w_{well-trained},h)$을 만들어낼 수 있다면,    
(학습에 쓰이지 않은) 임의의 새로운 데이터 $x_{new}$에 대한 정답값 $y(x_{new})$ 예측을,  
모형의 출력을 통해 매우 높은 정확도로 얻을 수 있게 된다.  

현재 수많은 분야에서 사용되는 인공지능 알고리즘들의 공통적인 목적은 (많은 경우) 정확한 예측을 위함이며,  
이러한 예측모형의 구현에 있어서, 데이터 기반의 통계학습모형 (기계학습모형)은 매우 높은 성능을 보이며,  
인공지능의 발전을 이끌고 있다.

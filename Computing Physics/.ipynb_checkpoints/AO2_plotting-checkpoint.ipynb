```python
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt #pyplot 를 plt 로 가져오기

N = 5 #fits 파일 갯수

data = ['' for i in range(N)]
for i in range(N):
    j = i+1
    name_str = 'obj'+str(j)+'.dred.dat'
    with open(name_str, 'r') as myfile:
        data[i] = myfile.readlines()

wavelength = []
flux = []
min_len = min([len(data[i]) for i in range(N)])
for i in range(min_len):
    l = 0.  #temporal wavelenght
    f = []  #temporal flux
    for j in range(N):
        temp = data[j][i].split('  ')
        if j == 0:
            l = float(temp[0])
        if float(temp[0]) > 6817 or float(temp[0]) < 6330:
            break
        f.append(float(temp[1]))
    if np.median(f) > 0:
        wavelength.append(l)
        flux.append(np.median(f))

fig = plt.figure(figsize=(8,6))
ax = plt.subplot()
ax.plot(wavelength,flux)
ax.set_xlabel("Wavelength")
ax.set_ylabel("Flux")
ax.set_yscale('log')
plt.show()
```

```python
num_bin = 3

w = 0.
f_bin = [0.,0.,0.]
bin_wavelength, bin_flux = [], []

for index in range(len(wavelength)):
    if index % 3 == 0:
        f_bin[0] = flux[index]
        continue
    elif index % 3 == 1:
        w = wavelength[index]
        f_bin[1] = flux[index]
        continue
    else:
        f_bin[2] = flux[index]
        bin_wavelength.append(w)
        bin_flux.append(np.median(f_bin))

print(len(bin_wavelength))
fig = plt.figure()
ax1 = plt.subplot()
ax1.plot(bin_wavelength,bin_flux)
ax1.set_xlabel("Wavelength")
ax1.set_ylabel("Flux")
ax1.set_yscale('log')
plt.show()
```

```python
from scipy.optimize import curve_fit
def func_line(x, m, a, b):
    return m*(x-a)+b

def func(x, a, b, c):
    return a*np.exp(-(x-b)**2/(2*c**2))

wavelength6563=[i-6563 for i in wavelength]
flux14=[i*10**14 for i in flux]
line_slope = (flux14[len(flux)-1]-flux14[0])/(wavelength6563[len(wavelength)-1]-wavelength6563[0])
cor_flux14=[flux14[i] - func_line(wavelength6563[i], line_slope, wavelength6563[0], flux14[0]) for i in range(len(flux))]

popt, pcov = curve_fit(func, wavelength6563, cor_flux14)
center_lambda = popt[1]+6563
x = [i+6328 for i in range(470)]
y = [(func_line(i-6563, line_slope, wavelength6563[0], flux14[0]) + func(i, popt[0], center_lambda, popt[2]))*10**(-14) for i in x]

FWHM = 2*math.sqrt(2*np.log(2))*popt[2] #FWHM=2*sqrt(2ln2)*sigma
velocity_FWHM = (FWHM/center_lambda)*300000
print(" center_wavelength\tFWHM\t\tvelocity_FWHM")
print(center_lambda, FWHM, velocity_FWHM)

ax2 = plt.subplot()
ax2.plot(wavelength,flux,label="Spectrum")
ax2.plot(x,y,label="Gaussian fitting")
ax2.set_title("""<$H\\alpha$ Emission Line Gaussian fitting>
$\lambda_{center}$ : 6540.43$\AA$, FWHM : 324.68$\AA$, $\sigma_{FWHM}$ : 14892km/s""")
ax2.set_xlabel("Wavelength")
ax2.set_ylabel("Flux")
ax2.set_yscale('log')
ax2.legend(loc='upper right')
plt.show()
```

```python
name_str1 = 'sn1993J_1995_UV_OPTICAL_FINAL.dat'
with open(name_str1, 'r') as myfile:
    data1 = myfile.readlines()

wavelength_sample = []
flux_sample = []
count = 0
for i in data1:
    count += 1
    
    if i[0] != ' ':
        temp = i.split('  ')
    else:
        temp = i[6:].split('  ')
    
    wl = float(temp[0])
    fl = float(temp[1])
    if wl < 5500 or wl > 10000:
        continue
    wavelength_sample.append(wl)
    flux_sample.append(fl*10)

data = ['' for i in range(N)]
for i in range(N):
    j = i+1
    name_str = 'obj'+str(j)+'.dred.dat'
    with open(name_str, 'r') as myfile:
        data[i] = myfile.readlines()

wavelength = []
flux = []
min_len = min([len(data[i]) for i in range(N)])
for i in range(min_len):
    l = 0.  #temporal wavelenght
    f = []  #temporal flux
    for j in range(N):
        temp = data[j][i].split('  ')
        if j == 0:
            l = float(temp[0])
        if float(temp[0]) > 8800 or float(temp[0]) < 5500:
            break
        f.append(float(temp[1]))
    if np.median(f) > 0:
        wavelength.append(l)
        flux.append(np.median(f))

fig = plt.figure()
ax4 = plt.subplot()
ax4.plot(wavelength,flux,label='Cas A')
ax4.plot(wavelength_sample,flux_sample,label='SN 1993J')
ax4.set_title("<Comparing Cas A to SN 1993J>")
ax4.set_xlabel("Wavelength")
ax4.set_ylabel("Flux")
ax4.set_yscale('log')

ax4.legend(loc='upper right')
plt.show()
```
